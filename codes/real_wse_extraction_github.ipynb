# Script for web scrapping cwc real time data
### import libraries 

```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import datetime
from datetime import date
from datetime import timedelta
import matplotlib.pyplot as plt
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
import os
```

### change as per requirements

```python
file_path_exp="F:\\phd_work\\cwc_real_time_data\\updated_real_time\\data_files\\2020_2024_July_31\\"

new_file_suffix='2020_2024_July.xlsx'
```

### data extracting and updating old files

```python
df_code=pd.read_excel('F:\\phd_work\\cwc_real_time_data\\updated_real_time\\data_files\\2020_updated\\name-code.xlsx',index_col=0)
url = "https://ffs.india-water.gov.in/web-api/getHGStationDataForFFS/"

datetime_format = '%Y-%m-%d %H:%M:%S.%f'


for name_station in df_code['name']:
    print('Data retrieving for ->',name_station)

    zz=df_code[df_code['name']==name_station].index.values
    index_name = zz[0] if len(zz) > 0 else None

    payload = {
        "stationCode": f"'{index_name}'",
        "startDate": "2020-01-01",
        "endDate": "2024-07-31"
    }
    response = requests.post(url, json=payload)

    if response.status_code == 200:
        try:
            data = response.json()
            print('Data retrived for ->',name_station)
            
            fl=[]
            df = pd.DataFrame()
            for j in data[:]:
                try:
                    ml=[j['stationCode'],datetime.datetime.strptime(j['actualTime'],datetime_format),name_station,j['value']]
                    fl.append(ml)
                except:
                    pass
            if fl:
                fl=sorted(fl)
                df=pd.DataFrame(fl)
                df.columns=['code','time','name','WSE']
                df.to_excel(file_path_exp+name_station+new_file_suffix) # new file end
                print('Data exported for station ',name_station)
            else:
                print('Data export failed for XXXXXXXXXXXXXXXXX',name_station)
                         
        except:
            print('Data not found for the station ',name_station)
    else:
        print("Error: Failed to retrieve data. Status code:", response.status_code)
        
```
