```python
df_code=pd.read_excel('F:\\phd_work\\cwc_real_time_data\\updated_real_time\\data_files\\2020_updated\\name-code.xlsx',index_col=0)
url = "https://ffs.india-water.gov.in/web-api/getHGStationDataForFFS/"
file_path_exp="F:\\phd_work\\cwc_real_time_data\\updated_real_time\\data_files\\2020_2024_June_25\\"
file_path_new="F:\\phd_work\\cwc_real_time_data\\updated_real_time\\data_files\\2020_2024_July_15\\"

old_file_suffix='2020_2024_6_25.xlsx'
new_file_suffix='2022_2024_7_15.xlsx'

datetime_format = '%Y-%m-%d %H:%M:%S.%f'


for name_station in df_code['name'][:100]:
    print('Data retrieving for ->',name_station)

    zz=df_code[df_code['name']==name_station].index.values
    index_name = zz[0] if len(zz) > 0 else None

    payload = {
        "stationCode": f"'{index_name}'",
        "startDate": "2024-06-26",
        "endDate": "2024-07-15"
    }
    response = requests.post(url, json=payload)

    if response.status_code == 200:
        try:
            data = response.json()
            print('Data retrived for ->',name_station)
            
            fl=[]
            df = pd.DataFrame()
            for j in data[:]:
                try:
                    ml=[j['stationCode'],datetime.datetime.strptime(j['actualTime'],datetime_format),name_station,j['value']]
                    fl.append(ml)
                except:
                    pass
            if fl:
                fl=sorted(fl)
                df=pd.DataFrame(fl)
                df.columns=['code','time','name','WSE']
                old_file=pd.read_excel(file_path_exp+name_station+old_file_suffix) # update old file end
                df_z = pd.concat([old_file, df], ignore_index=True)
                df_z = df_z.drop(columns=['Unnamed: 0'])
                df_z.to_excel(file_path_new+name_station+new_file_suffix) # new file end
                print('Data exported for station ',name_station)
            else:
                print('Data export failed for XXXXXXXXXXXXXXXXX',name_station)
                df_z=old_file
                df_z.to_excel(file_path_new+name_station+new_file_suffix)
           
        except:
            print('Data not found for the station ',name_station)
    else:
        print("Error: Failed to retrieve data. Status code:", response.status_code)
        
```
